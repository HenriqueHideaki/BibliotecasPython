```python
# Exemplo: contar a frequência de ocorrência de cada categoria em uma variável categórica
df['tipo_acidente'].value_counts()

plt.figure(figsize=(14, 6))
acidentes_por_tipo = df['tipo_acidente'].value_counts().sort_values(ascending=False)
sns.barplot(x=acidentes_por_tipo.index, y=acidentes_por_tipo.values)
plt.xticks(rotation=45)
plt.xlabel('Tipo de Acidente')
plt.ylabel('Contagem')
plt.title('Distribuição dos Tipos de Acidente')
plt.tight_layout()
plt.show()
```

```python
df['Data/Hora'] = pd.to_datetime(df['horario'])
# Extrair a hora e o minuto de cada registro
df['Hora'] = df['Data/Hora'].dt.hour
df['Minuto'] = df['Data/Hora'].dt.minute
df['Hora'] = ((df['Hora'] * 60 + df['Minuto']) // 30) * 30 // 60
# df['Hora'] = (df['Hora'] // 2) * 2
contagem_acidentes = df['Hora'].value_counts().sort_index()
horas_formatadas = contagem_acidentes.index.map(lambda x: f"{x:02d}:00 {'AM' if x < 12 else 'PM'}")
contagem_formatada = pd.DataFrame({'Hora': horas_formatadas, 'Contagem': contagem_acidentes.values})
print(contagem_formatada)



plt.figure(figsize=(24, 8))
for i, contagem in enumerate(contagem_formatada['Contagem']):
    plt.text(contagem_formatada['Hora'][i], contagem, str(contagem), ha='center', va='bottom')
plt.plot(contagem_formatada['Hora'], contagem_formatada['Contagem'], marker='o')


# Adicionar rótulos e título ao gráfico
plt.xlabel('Hora')
plt.ylabel('Número de Acidentes')
plt.title('Contagem de Acidentes por Hora')

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Verificar se há informações sobre condições climáticas no dataset

df_amanhecer = df[df['fase_dia'] == 'Amanhecer']

# Criar um DataFrame com as variáveis relevantes (exemplo)
heatmap_data = df_amanhecer[['condicao_metereologica', 'tipo_acidente']]

# Contar a ocorrência de cada combinação de condições climáticas e tipo de acidente
contagem_acidentes = heatmap_data.groupby(['condicao_metereologica', 'tipo_acidente']).size().unstack(fill_value=0)

# Criar o mapa de calor com Seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(contagem_acidentes, cmap='YlOrRd', annot=True, fmt='d')

# Adicionar rótulos e título ao gráfico
plt.xlabel('Tipo de Acidente')
plt.ylabel('Condições Climáticas')
plt.title('Contagem de Acidentes por Condições Climáticas e Tipo de Acidente ao Amanhecer')

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap


# Filtrar os acidentes ocorridos na UF RN
data_rn = df[df['uf'] == 'RN']
data_rn_amanhecer = data_rn[data_rn['fase_dia'] == 'Amanhecer']
# Selecionar as colunas de latitude e longitude
lat_lng_data = data_rn_amanhecer[['latitude', 'longitude']]

# Criar o mapa centrado na latitude e longitude médias
mapa = folium.Map(location=[lat_lng_data['latitude'].mean(), lat_lng_data['longitude'].mean()], zoom_start=10)

# Converter as coordenadas para uma lista de tuplas
coordenadas = list(zip(lat_lng_data['latitude'], lat_lng_data['longitude']))

# Criar o mapa de calor com base nas coordenadas
HeatMap(coordenadas).add_to(mapa)

# Exibir o mapa
mapa
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Agrupar os dados por condição climática e contar a quantidade de acidentes
grouped_data = df.groupby('condicao_metereologica').size()

# Calcular média e desvio padrão dos acidentes
media = grouped_data.mean()
desvio_padrao = grouped_data.std()

# Gerar uma amostra aleatória de valores seguindo uma distribuição normal
dados = np.random.normal(media, desvio_padrao, 1000)

# Plotar o histograma dos dados
plt.hist(dados, bins=30, density=True, alpha=0.7)

# Gerar os valores para a curva da distribuição gaussiana
x = np.linspace(min(dados), max(dados), 100)
pdf = norm.pdf(x, media, desvio_padrao)

# Plotar a curva da distribuição gaussiana
plt.plot(x, pdf, color='red')

# Configurar o título e os rótulos dos eixos
plt.title('Distribuição Gaussiana - Acidentes de Trânsito por Condição Climática')
plt.xlabel('Quantidade de Acidentes')
plt.ylabel('Densidade')

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Filtrar a coluna relevante
data_filtered = df['condicao_metereologica']

# Calcular a frequência de cada condição meteorológica
frequencies = data_filtered.value_counts()

# Normalizar as frequências para obter as probabilidades
probabilities = frequencies / frequencies.sum()

# Criar a distribuição de probabilidades
distribution = pd.DataFrame({'condicao_meteorologica': probabilities.index, 'probabilidade': probabilities.values})

# Ordenar a distribuição por ordem decrescente de probabilidades
distribution = distribution.sort_values(by='probabilidade', ascending=False)

# Visualizar a distribuição de probabilidades em um gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(distribution['condicao_meteorologica'], distribution['probabilidade'])
plt.xlabel('Condição Meteorológica')
plt.ylabel('Probabilidade')
plt.title('Condição Meteorológica')
plt.xticks(rotation=45)
plt.show()
```
